```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for the bar chart
x = ['Wave', 'Zero Crossing Rate', 'Mel Spectrogram', 'MFCC plots']
y1 = [0.48, 0.5, 0.84, 0.7]
y2 = [0.65, 0.62, 0.74, 0.68]
y3 = [0.42, 0.42, 0.42, 0.42]

# Set the width of each bar
bar_width = 0.25

# Calculate the x position of each group of bars
x_pos = np.arange(len(x))

# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6)) # Set figure size
ax.bar(x_pos - bar_width, y1, width=bar_width, label='Baseline-images', color='#E5E0FF')
ax.bar(x_pos, y2, width=bar_width, label='Multi-modal Network', color='#7286D3')
ax.bar(x_pos + bar_width, y3, width=bar_width, label='Baseline-MFCC time series', color='#8EA7E9')

# Add title and axis labels
# ax.set_title("Accuracy Comparison between Baseline and Multi-modal Network")
ax.set_xlabel("Image Forms")
ax.set_ylabel("Prediction Accuracy")

# Set x-axis labels
ax.set_xticks(x_pos)
ax.set_xticklabels(x)

# Add legend and grid
ax.legend()
ax.grid(False)

# Show only bottom and left border
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Display the plot
# plt.show()
plt.savefig('multimodal_accuracy.pdf')
```
